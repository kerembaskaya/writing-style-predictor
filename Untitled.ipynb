```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas

surtunme = 0.1 #newton
hedef = 20 #metre / saniye
zaman = 0
son_zaman = 15
timestep = 0.01
dizi_boyutu = int(son_zaman / timestep)
agirlik = 2 #kg
guc = 0 #newtons
maks_guc = 10 #newtons
hiz = 0
toplam_timestep = 0

hizlar = np.zeros(dizi_boyutu)
gucler = np.zeros(dizi_boyutu)
zamanlar = np.zeros(dizi_boyutu)

kp = 1.5
ki = 0.01
kd = 0.052

# kp = 0.9
# ki = 0.005
# kd = 0.0042


class PID(object):
    def __init__(self,kp,ki,kd, hedef):
        self.kp = kp
        self.ki = ki
        self.kd = kd
        self.setpoint = hedef
        self.hata = 0
        self.integral_hata = 0
        self.son_hata = 0
        self.turev_hata = 0
        self.cikis = 0
    def hesaplar(self, hiz):
        self.hata = hedef - hiz
        self.integral_hata += self.hata * timestep
        self.turev_hata = (self.hata - self.son_hata) / timestep
        self.son_hata = self.hata
        self.cikis = self.kp*self.hata + self.ki*self.integral_hata + self.kd*self.turev_hata

        if self.cikis > maks_guc:
            self.cikis = maks_guc
        elif self.cikis < -maks_guc:
            self.cikis = -maks_guc



        return self.cikis

pid = PID(kp, ki, kd, hedef)

for i in range(dizi_boyutu):

    guc = pid.hesaplar(hiz)
    total_guc = guc - surtunme
    if total_guc < 0 and hiz <= 0:
        total_guc = 0 #surtunmeden dolayi arac geri gitmesin diye


    gucler[i] = guc
    zamanlar[i] = toplam_timestep
#     print("Guc", guc) # test için
    #time.sleep(timestep) #test icin
    zaman += 1
    hiz += (total_guc / agirlik) * timestep
    hizlar[i] = hiz
    toplam_timestep += timestep

print(len(hizlar), hizlar[-1], zamanlar[-1], max(hizlar), max(zamanlar), max(gucler))
print(f"First: {gucler[0]}, Second: {gucler[1]}")


def plot():
    fig, axes = plt.subplots(1, 2, figsize=(15,5))


    d = {"velocity": hizlar, "time": zamanlar, "power": gucler}
    df = pandas.DataFrame(d)

    # Hız
    ax = axes[0]
    sns.lineplot(x="time", y="velocity", data=df, ax = ax)
    ax.set_title("PID " + "Kp: " +str(kp) + " Ki: " +str(ki) + " Kd: " + str(kd))
    ax.set_ylabel("hiz (m/s)")
    ax.set_xlabel("saniye")
    
    constant = pandas.DataFrame({"time": list(range(son_zaman + 1)), "velocity": [hedef] * (son_zaman + 1)})
    sns.lineplot(x="time", y="velocity", data = constant, ax=ax)

    # Guc
    ax = axes[1]
    sns.lineplot(x="time", y="power", data=df, ax = ax)

    ax.set_title("Guc - Zaman")
    ax.set_ylabel("Guc (newton)")
    ax.set_xlabel("saniye")

plot()


plt.show()


```
